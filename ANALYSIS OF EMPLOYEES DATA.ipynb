```python
# HOW DOES THE DATA VARY IN TERMS OF GENDER
gender=df['Gender'].value_counts()
gender
```

```python
plt.pie(gender,labels=gender.index,autopct='%.2f%%',startangle=140 )
plt.title('Gender Distribution')
```

```python
#grouping of gender with department column
gender_dep=df.groupby('Gender')['Department'].value_counts()
gender_dep
```

```python
#grouping of gender by marital status
gender_m_status=df.groupby('Gender')['MaritalStatus'].value_counts()
gender_m_status
```

```python
# what is the travelling distribution in the firm
bs_travel=df['BusinessTravel'].value_counts()
bs_travel
```

```python
#pie chart of the business travels distribution
plt.figure(figsize=(8,6))
plt.pie(bs_travel,labels=bs_travel.index, autopct='%1.1f%%',startangle=140)
plt.title('business travel distribution')
```

```python
# individuals who have never had a business travel in the firm
no_travel=df.groupby('BusinessTravel').get_group('No Travel ')
no_travel
```

```python
#department distribution
dep_dist=df['Department'].value_counts()
dep_dist
```

```python
# bar graph of department distribution
data={'Department': ['Technology','Sales','Human Resources'],
      'count':[961,446,63]}
sns.set(style="whitegrid")
plt.figure(figsize=(8,5))
sns.barplot(data=data, x='Department', y='count', color='skyblue')
plt.xlabel('Department')
plt.ylabel('count')
plt.title('department distribution bar graph')
plt.show
```
