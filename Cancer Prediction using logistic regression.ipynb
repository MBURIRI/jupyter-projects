```python
df=df.drop(['Unnamed: 32','id'], axis=1)
df.head()
```

```python
from sklearn import preprocessing
#initialize label encorder\
label_encoder=preprocessing.LabelEncoder()
# convert diagnosis variable to nueric
df['diagnosis']=label_encoder.fit_transform(df['diagnosis'])
df.head(200)
```

```python
# Create the plot
ax = df['diagnosis'].value_counts().plot(kind='bar', color=['blue', 'orange'], figsize=(6, 4), edgecolor='black')

# Add annotations
for bar in ax.patches:
    ax.annotate(f'{bar.get_height()}', 
                (bar.get_x() + bar.get_width() / 2, bar.get_height()), 
                ha='center', va='bottom', fontsize=12, fontweight='bold', color='black')

# Customize appearance
plt.title("Distribution of Diagnosis", fontsize=14, fontweight='bold')
plt.xlabel("Diagnosis", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.xticks(rotation=0)  # Keep x-axis labels horizontal
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Add gridlines for readability
plt.show()
```

```python
# divide the dataset into dependent and independent variable
y=df['diagnosis']
x=df.drop(['diagnosis'],axis=1)
```

## normalizing the data

```python
from sklearn.preprocessing import StandardScaler
#create a scaler object
scaler= StandardScaler()
#fit the scaler to the data and tranform it
x_scaled=scaler.fit_transform(x)
x_scaled
```

## split the data

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test=train_test_split(x_scaled, y, test_size=0.3, random_state=42)
```

## train the model

```python
from sklearn.linear_model import LogisticRegression
# create the logistic regression(lr)
lr=LogisticRegression()
# fit the model
lr.fit(x_train,y_train)
# predict the dependent variable on the test data
y_pred=lr.predict(x_test)
y_pred
```

## Evaluation of the model

```python
from sklearn.metrics import accuracy_score

accuracy=accuracy_score(y_test,y_pred)
print(f'Accuracy score: {accuracy:.2f}')
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
